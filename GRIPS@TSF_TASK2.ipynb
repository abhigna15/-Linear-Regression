**Import Dataset**

```
import pandas as pd 
```

```
df = pd.read_csv("/content/drive/My Drive/linear_Reg.csv")
```

```
df.head()
```

```
df.info()
```

```
X = df[['Hours']]
type(X)
```

```
y= df[['Scores']]
type(y)
```

**Train Test and Split**

```
from sklearn.model_selection import train_test_split
```

```
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= 0.25, random_state =0)
```

**Import Linear Regression**

```
from sklearn.linear_model import LinearRegression
```

```
linear = LinearRegression()
```

```
#X = X.reshape(-1,1)
linear.fit(X_train,y_train)
linear.score(X_train,y_train)
```

```
# Plotting line graph
import matplotlib.pyplot as plt
# For a straight line y = mx +c, 
#here m is weights of feature variables and x,y are the values in the table
line = linear.coef_*X+linear.intercept_
plt.scatter(X,y)
plt.plot(X,line)
```

```
data = pd.read_csv("/content/drive/My Drive/linear_reg_predict_1.csv")
y_pred=linear.predict(data)
y_pred
```

```
data1 = [[9.25]]
print(linear.predict(data1),"Is the score, student will get after studying for 9.25 hours")
```

**Evaluating the model**

```
from sklearn.metrics import mean_absolute_error
import math as mt
error = mean_absolute_error(y_test,y_pred)
print(mt.sqrt(error))
```
